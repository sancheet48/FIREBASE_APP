```python
# pip install spacy , nltk
# python -m spacy download en_core_web_sm

import nltk
import spacy
from nltk.tokenize import word_tokenize
from nltk.stem import PorterStemmer, WordNetLemmatizer
from nltk.corpus import stopwords
from nltk.chunk import RegexpParser
from nltk.tag import pos_tag
import en_core_web_sm

# Initialize NLTK and spaCy
nltk.download('punkt')
nltk.download('wordnet')
nltk.download('stopwords')
nltk.download('averaged_perceptron_tagger')
nlp = spacy.load('en_core_web_sm')

# Input text
input_text = "Find out all the rows where sex is given as male and pclass is referred as 1"

# Word Tokenization
tokens = word_tokenize(input_text)

# Stemming
stemmer = PorterStemmer()
stemmed_tokens = [stemmer.stem(token) for token in tokens]

# Lemmatization
lemmatizer = WordNetLemmatizer()
lemmatized_tokens = [lemmatizer.lemmatize(token) for token in tokens]

# Identifying Stop Words
stop_words = set(stopwords.words('english'))
filtered_tokens = [token for token in tokens if token.lower() not in stop_words]

# Dependency Parsing, POS tags, Named Entity Recognition (NER), and Chunking using spaCy
doc = nlp(input_text)

# Dependency Parsing
dependency_tree = [(token.text, token.dep_, token.head.text) for token in doc]

# POS tags
pos_tags = [(token.text, token.pos_) for token in doc]

# Named Entity Recognition (NER)
ner_tags = [(ent.text, ent.label_) for ent in doc.ents]

# Chunking using RegexpParser (You can define your custom rules)
custom_chunk_rules = r"""Chunk: {<DT>?<JJ>*<NN.*>+}"""
chunk_parser = RegexpParser(custom_chunk_rules)
chunked_tree = chunk_parser.parse(pos_tag(filtered_tokens))

# Print the results
print("Word Tokenization:", tokens)
print("Stemming:", stemmed_tokens)
print("Lemmatization:", lemmatized_tokens)
print("After Removing Stop Words:", filtered_tokens)
print("Dependency Parsing:", dependency_tree)
print("POS tags:", pos_tags)
print("Named Entity Recognition (NER):", ner_tags)
print("Chunking:", chunked_tree)
```

```python
# Dependency parsing result
dependency_parsing_result = [
    ('Find', 'ROOT', 'Find'),
    ('out', 'prt', 'Find'),
    ('all', 'predet', 'rows'),
    ('the', 'det', 'rows'),
    ('rows', 'nsubjpass', 'referred'),
    ('where', 'advmod', 'given'),
    ('sex', 'nsubjpass', 'given'),
    ('is', 'auxpass', 'given'),
    ('given', 'relcl', 'rows'),
    ('as', 'prep', 'given'),
    ('male', 'pobj', 'as'),
    ('and', 'cc', 'male'),
    ('pclass', 'conj', 'male'),
    ('is', 'auxpass', 'referred'),
    ('referred', 'ccomp', 'Find'),
    ('as', 'prep', 'referred'),
    ('1', 'pobj', 'as')
]

# Initialize variables to store inferred conditions
sex_condition = None
pclass_condition = None

# Iterate through the parsing result to extract conditions
for token, relation, head in dependency_parsing_result:
    if relation == 'nsubjpass' and token == 'sex':
        # "sex" is associated with the passive subject, indicating a condition on sex
        sex_condition = head
    elif relation == 'conj' and head == 'male':
        # "pclass" is connected to "male" through "conj," indicating a condition on pclass
        pclass_condition = token

# Print the inferred conditions
if sex_condition is not None:
    print(f"Condition on 'sex': {sex_condition}")
if pclass_condition is not None:
    print(f"Condition on 'pclass': {pclass_condition}")
```
